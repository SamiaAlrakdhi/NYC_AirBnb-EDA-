# Calculate listings for each host

```python
########################
bighost = df.groupby(["host_name","city"]).agg({'calculated_host_listings_count': 'sum'}).sort_values('calculated_host_listings_count',ascending=False)
bighost.reset_index(inplace=True)
bighost.head(10).plot(kind="bar",x='host_name',figsize =(12,8))
plt.xticks(rotation=-45);
```

# Compare the number of room types in each city

```python
plt.figure(figsize=(15,6))
sns.countplot(data=df,x='city',hue='room_type',palette='Blues')
plt.title('City with room types they alot')
plt.show()
```
